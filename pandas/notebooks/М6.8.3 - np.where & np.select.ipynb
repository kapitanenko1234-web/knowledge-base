# Функции WHERE & SELECT библиотеки numpy

# Изменение DataFrame по нескольким условиям

```python
df = pd.read_csv('files/sample_1.csv')
df
```

## Базовое использование np.where

```python
# пример работы функции np.where (Аналог функции 'ЕСЛИ' в Excel)

df['result']=np.where(df['point']>59, 'Passed', 'Failed')
df
```

## Другие примеры использования np.where

### Пример 1 - Получить индексы строк, удовлетворяющих условию 'возраст < 30'

```python
inds = np.where(df.age<30)
display(type(inds),inds)
```

### Пример 2 - Получить данные из конкретного столбца по определенному условию, а если условие не выполняется, получить данные из другой колонки

```python
result=np.where(df.age<30,  # условие (возраст < 30 лет)
         df.name,    # если условие выполняется, получить имя
         df.point)   # если условие не выполняется, получить результат тестирования
display(type(result), result)
```

## Несколько условий. Функция np.select

```python
df2 = pd.read_csv('files/sample_1.csv')
df2
```

### Два условия: если участник прибыл из штата Нью-йорк, мы от полученного результата отнимаем один балл, а если участник - из штата Калифорния, то, наоборот, один бал надо прибавить.

```python
condlist = [(df2['state']=='NY'),
             (df2['state']=='CA')]
choicelist = [df2['point']-1,
               df2['point']+1]
```

```python
# по умолчанию функция select поставит ноль, если не находит соответствие условию (default=0)

df2['new_point'] = np.select(condlist, choicelist)
df2
```

```python
# указываем в аргументе default колонку с изначальными результатами

df2['new_point'] = np.select(condlist, choicelist, default=df2['point'])
df2
```
