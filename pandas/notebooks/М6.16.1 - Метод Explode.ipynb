```python
import pandas as pd
import numpy as np
print('Версия pandas:', pd.__version__)
```

```python
df = pd.DataFrame({'English': [[78, 85], np.nan, 70, [75, 78]],
                   'French': [[69, 75], np.nan, 69, [73, 80]],
                   'Spanish': [[68, 78], np.nan, 67, [79, 85]]}, index=[1001, 1002, 1003, 1004])
df.index.name='student_id'
df
```

```python
# разделяем скученные данные из одной колонки
# передаем имя нужной колонки
df.explode('English')
```

```python
# разделяем скученные данные из нескольких колонок
# передаем нужные колонки в списке
df.explode(['English','French','Spanish'])
```
