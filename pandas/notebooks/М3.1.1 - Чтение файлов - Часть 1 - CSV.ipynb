# Формат CSV

### CSV - Comma-Separated Values - значения, разделённые запятыми

```python
# вот так выглядит csv-файл "расписание", если открыть его через программу "Блокнот"
# Строка таблицы соответствует строке текста, которая содержит одно или несколько полей, разделенных запятыми.
display(Image(filename='files/скрин_расписание.png'))
```

# Метод для чтения файла .read_csv()

```python
# читаем файл в переменную df1 с помощью метода .read_csv()
# по умолчанию параметр sep=','
df1=pd.read_csv('files/расписание')
df1
```

# Метод для записи файла .to_csv()

```python
# транспонируем датафрейм и сохраняем его в переменную df2
df2=df1.T
# даем названия колонкам датафрейма df2
df2.columns=['8.30-9.15','9.25-10.10', '10.30-11.15','11.35-12.20','12.35-13.20', '13.30-14.15']
df2
```

```python
# записываем получившийся датафрейм в csv-файл в папку files под названием "расписание2"
df2.to_csv('files/расписание2')
```

```python
# вот так выглядит csv-файл "расписание2", если открыть его через программу "Блокнот"
display(Image(filename='files/скрин_расписание2.png'))
```

```python
# при открытии сохраненного csv-файла "расписание2" мы обнаруживаем, что колонка с индексами строк стала колонкой "Unnamed: 0"
df3=pd.read_csv('files/расписание2')
df3
```

```python
# чтобы вернуть индексы на прежнее место, при окрытии файла передаем в параметр index_col список с позицией нужной колонки -"0"
df3=pd.read_csv('files/расписание2', index_col=[0])
df3
```

```python
# вот так выглядит csv-файл "оценки_Абрамов", если открыть его через программу "Блокнот"
display(Image(filename='files/скрин_оценки_Абрамов.png'))
```

```python
# первая строчка файла становится header по умолчанию
df4_1=pd.read_csv('files/оценки_Абрамов')
df4_1
```

```python
# Если в файле нет заголовков колонок, то параметр header указываем как None
df4_2=pd.read_csv('files/оценки_Абрамов', header=None)
df4_2
```

```python
# вот так выглядит csv-файл "оценки", если открыть его через программу "Блокнот"
display(Image(filename='files/скрин_оценки.png'))
```

```python
# с помощью параметра skiprows убираем первые три строчки из вывода
# Если нужно исключить первые три строки, то необходимо писать skiprows = 3
# Если нужно удалить только третью строку, то пишем skiprows = [3].
# с помощью параметра names задаем названия столбцов
df6=pd.read_csv('files/оценки', skiprows=3, names=['ФИО студента', 'сентябрь','октябрь','ноябрь','декабрь'])
df6
```
