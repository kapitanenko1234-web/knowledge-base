```python
df_students = df_students.dropna()
```

```python
df_students['Exam_Score'].std()
```

```python
q1 = df_students['Exam_Score'].quantile(0.25)
q3 = df_students['Exam_Score'].quantile(0.75)
iqr = q3 - q1

condition=(df_students['Exam_Score'] > q1 - 1.5 * iqr) & (df_students['Exam_Score'] < q3 + 1.5 * iqr)

df_students_new = df_students.loc[condition]
df_students_new.reset_index(drop=True, inplace=True)
```

```python
df_students_new.info()
```

```python

cat=[0,64,70,df_students_new['Exam_Score'].max()]

df_students_new['Score_type']= pd.cut(df_students_new['Exam_Score'], cat, 
    labels=['Low (<65)', 'Middle (65-70)', 'High (>70)'])
```

```python
df_crosstab = pd.crosstab(df_students_new['Teacher_Quality'], df_students_new['Motivation_Level'])
df_crosstab
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(df_crosstab)
chi2, p, dof, expected = chi2_contingency(df_crosstab)
# из полученных значений нам нужно только значение p
print(f"p-значение: {p}")
```
