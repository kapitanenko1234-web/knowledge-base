# Метод filter() - извлечение строк/столбцов из DataFrame в соответствии с метками

### Условия для фильтрации можно задавать с помощью параметров:

# items | like | regex

### Необходимо выбрать один параметр из трех, иначе возникнет ошибка

### Параметр axis определяет ось для фильтрации

# items

Параметр items принимает список, содержащий одно или несколько имен столбцов/строк. При его указании функция возвращает новый датафрейм, содержащий только те столбцы/строки, которые указаны в списке.

# like

Параметр like принимает строковое значение и возвращает датафрейм, содержащий только те столбцы/строки, в названии которых присутствует данная подстрока.

# regex

Параметр regex аналогичен like, но вместо возврата имен столбцов/строк, содержащих подстроку, он возвращает те столбцы/строки, имена которых соответствуют регулярному выражению Python.

# axis

Параметр axis позволяет указать, как именно следует фильтровать датафрейм - по именам столбцов или по именам/индексам строк.

#### Пример 1

```python
df.filter(items=['A', 'C'])
```

#### Пример 2

```python
# параметры items и like указаны одновременно, поэтому возникает ошибка
df.filter(items=['A', 'C'], like='A')
# TypeError: Keyword arguments `items`, `like`, or `regex` are mutually exclusive
```

#### Пример 3.1

```python
# если ни одна строка или столбец не удовлетворяет условию, то возвращается пустой датафрейм.
# ищем колонку с меткой 'X'
df.filter(items=['X'])
```

#### Пример 3.2

```python
# ищем колонку с меткой 'X' + используем функцию print
print(df.filter(items=['X']))
```

#### Пример 3.3

```python
# если ни одна строка или столбец не удовлетворяет условию, то возвращается пустой датафрейм.
# ищем строку с меткой 'orange'
# указываем параметр axis=0, чтобы поиск производился по меткам строк
df.filter(items=['orange'], axis=0)
```

#### Пример 3.4

```python
# ищем строку с меткой 'orange' + используем функцию print
print(df.filter(items=['orange'], axis=0))
```

# Параметр axis

### Для фильтрации по строкам axis=0 или axis='index'
### Для фильтрации по столбцам axis=1 или axis='columns' или ничего не указываем

```python
print('Результат фильтрации по строкам:', 
      df.filter(items=['apple', 'pineapple'], axis=0),
      df.filter(items=['apple', 'pineapple'], axis='index'),
      
      'Результат фильтрации по столбцам:',
      df.filter(items=['A', 'C'], axis=1),
      df.filter(items=['A', 'C'], axis='columns'),
      
      sep="\n\n")
```

```python
# Если значение axis опущено, то фильтруются столбцы

df.filter(items=['A', 'C'])
```

```python
# одновременная фильтрация строк и столбцов невозможна
# если необходимо отфильтровать и строки, и столбцы, то просто повторите filter:
df.filter(items=['A', 'C']).filter(items=['apple', 'pineapple'], axis=0)
```

# Для точного совпадения используем параметр items

```python
df.filter(items=['A', 'C'])
```

```python
print(df[['C', 'A']],
      df.loc[['pineapple', 'apple']],
      sep="\n\n")
```

# Для частичного совпадения используем параметр like

```python
df.filter(like='ban', axis=0)
```

```python
df.filter(like='apple', axis=0)
```

# Для поиска по шаблону регулярного выражения используем regex

```python
# извлекаем строки, имена которых заканчиваются на 'a'
df.filter(regex='a$', axis=0)
```

```python
# извлекаем строки, имена которых начинаются с 'a' или 'b'
df.filter(regex='^(a|b)', axis=0)
```

```python
# извлекаем строки, имена которых содержат 'na' или 'ne'
df.filter(regex='(na|ne)', axis=0)
```

# Использование метода filter() c Series

```python
s = pd.Series([0, 1, 2], index=['apple', 'banana', 'pineapple'])
s
```

```python
print(s.filter(items=['pineapple', 'banana']),
      s.filter(like='apple'),
      s.filter(regex='^(a|b)'), sep="\n\n")
```
