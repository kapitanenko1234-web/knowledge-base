## ЧАСТЬ 1

### Cоздание DataFrame с использованием нескольких Series в качестве столбцов

```python
#Создадим три серии
name = pd.Series(['Анна', 'Екатерина', 'Елена', 'Светлана', 'Дарья'])
age = pd.Series([34, 25, 21, 43, 30])
zodiac = pd.Series(["Скорпион", "Рыбы", "Дева", "Водолей", "Весы"])
```

##### ПРИМЕР 1.1

```python
#создадим по одному датафрейму из каждого объекта series
name_df1 = name.to_frame(name='Имя')
age_df1 = age.to_frame(name='Возраст')
zodiac_df1 = zodiac.to_frame(name='Знак зодиака')

#объеденим три датафрейма в один
df_girls = pd.concat([name_df1, age_df1, zodiac_df1], axis= 1 )

#смотрим на итоговый датафрейм
df_girls
```

##### ПРИМЕР 1.2

```python
#создадим по одному датафрейму из каждого объекта series
name_df2 = name.to_frame(name='Имя')
age_df2 = age.to_frame(name='Возраст')
zodiac_df2 = zodiac.to_frame(name='Знак зодиака')

#объеденим три датафрейма в один (axis=0)
df_girls1_2 = pd.concat([name_df2, age_df2, zodiac_df2])

#посмотрим на итоговый датафрейм
df_girls1_2
```

##### ПРИМЕР 1.3

```python
#создадим по одному датафрейму из каждого объекта series
name_df3 = name.to_frame()
age_df3 = age.to_frame()
zodiac_df3 = zodiac.to_frame()

#объеденим три датафрейма в один (axis=0)
df_girls1_3 = pd.concat([name_df3, age_df3, zodiac_df3])

#посмотрим на итоговый датафрейм
df_girls1_3
```

##### ПРИМЕР 1.4

```python
# Создадим еще три серии c данными парней
name2 = pd.Series(['Михаил', 'Константин', 'Николай', 'Сергей', 'Александр'])
age2 = pd.Series([32, 20, 37, 50, 19])
zodiac2 = pd.Series(["Рак", "Козерог", "Лев", "Овен", "Близнецы"])
# создадим одному датафрейму из каждого объекта series
name_df4 = name2.to_frame(name='Имя')
age_df4 = age2.to_frame(name='Возраст')
zodiac_df4 = zodiac2.to_frame(name='Знак зодиака')

#объеденим три новых датафрейма в один
df_boys = pd.concat([name_df4, age_df4, zodiac_df4], axis= 1 )

#посмотрим на итоговый датафрейм
df_boys
```

##### Пример 1.5

```python
df_all = pd.concat([df_girls, df_boys], ignore_index=True)
df_all
```

## ЧАСТЬ 2

### Cоздание DataFrame с использованием нескольких Series в качестве строк

##### Пример 2.1

```python
# Создадим пять серий
row1 = pd.Series(['Аня', 34, "Скорпион"])
row2 = pd.Series(['Катя', 25, "Рыбы"])
row3 = pd.Series(['Лена', 21, "Дева"])
row4 = pd.Series(['Света', 43, "Водолей"])
row5 = pd.Series(['Даша', 30, "Весы"])
```

```python
#Создаем датафрейм, используя серии как строки
df2 = pd.DataFrame([row1, row2, row3, row4, row5])

#создаем названия колонок нашего датафрейма
df2.columns = ['Имя', 'Возраст', 'Знак Зодиака']

#смотрим на результат
df2
```

##### Пример 2.2

```python
df2.T
```

##### Пример 2.3

```python
df_ = pd.concat([row1, row2, row3, row4, row5], axis=1)
df_.index = ['Имя', 'Возраст', 'Знак Зодиака']
df_
```
