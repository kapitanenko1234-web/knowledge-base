## Метод .xs( ) cross section

    Синтаксис: dataframe.xs(key, axis, level, drop_level)
    
    Параметры метода:

    key - метка индекса, либо кортеж меток
    Обязательный параметр. Метка индекса строки, которую вы хотите вернуть. 
    Используйте кортеж меток, чтобы вернуть более одной метки

    axis - ось (строк или столбцов)
    Указывать необязательно. По умолчанию 0.

    level - Указываем, какие уровни мы используем в своем запросе. Уровни могут быть указаны по метке или позиции.

    drop_level - Указываем, возвращать ли объекты того же уровня, что и сам аргумент key, или нет
    По умолчанию True. 

## Пример 4.1

```python
df_w.xs(key='CA Warehouses')
```

## Пример 4.2

```python
df_w.xs(key='CA Warehouses', drop_level=False)
```

## Пример 4.3

```python
# используем кортеж меток
# на выходе получаем все имеющиеся данные по Сан-Диего 
df_w.xs(key=('CA Warehouses','San Diego'))
```

## Пример 4.4

```python
# Мы не можем указать штат и тип продукции, не указав город или используемые уровни
try:
    df_w.xs(key=('CA Warehouses','AC'))
except KeyError as e:
     print("KeyError:", e)
```

## Пример 4.5

```python
# указываем нужные уровни в аргументе level
# чтобы получить данные по Калифорнийским складам в части только продукции AC, нам надо указать уровни 'State'и'Product'
# по умолчанию drop_level=True, то есть запрошенные уровни скрыты

df_w.xs(('CA Warehouses', 'AC'), level=('State', 'Product'))

# другие варианты написания этого же запроса:
# df_w.xs(('CA Warehouses', 'AC'), level=['State', 'Product'])
# df_w.xs(('CA Warehouses', 'AC'), level=(0, 2))
# df_w.xs(('CA Warehouses', 'AC'), level=[0, 2])
```

```python
# прописываем drop_level=False, чтобы видеть метки запрошенных уровней 'State'и'Product'

df_w.xs(('CA Warehouses', 'AC'), level=('State', 'Product'), drop_level=False)
```

## Пример 4.6

```python
# Мы не можем указать одиночную метку ненулевого уровня (например, TV), не указав уровень в параметре level
try:
    df_w.xs(key='TV')
except KeyError as e:
     print("KeyError:", e)
```

## Пример 4.7

```python
# здесь мы запросили все данные по телевизорам по всем складам во всех имеющихся шатах
df_w.xs(key='TV', level='Product')
```

## Пример 4.8

```python
# здесь мы сделали рассчет количества единиц товара категории AC по всем складам в Калифорнии
# сначала по каждому отчетному периоду
# в итоге вывели общее количество едениц товара категории AC за все время

display(df_w.xs(('CA Warehouses', 'AC'), level=('State', 'Product'), drop_level=False),
        df_w.xs(('CA Warehouses', 'AC'), level=('State', 'Product'), drop_level=False).sum(),
        df_w.xs(('CA Warehouses', 'AC'), level=('State', 'Product'), drop_level=False).sum().sum())
```

## Пример 4.9

```python
# делаем поперечный срез другой оси, указываем axis=1

df_w.xs('Jan-Jun', axis=1, level='Semester', drop_level=False)
```

# pandas.IndexSlice

    pd.IndexSlice - специальный объект для более удобного выполнения многоиндексной нарезки.

```python
# создаем спец.объект для многоиндексной нарезки
i_sl = pd.IndexSlice
```

## Пример 7.1

```python
# Нам нужны все штаты и города, а из продукции мы выбираем только AC и TV
df_w.loc[i_sl[:, :, ['AC', 'TV']], :]
```

## Пример 7.2

```python
# можно выполнять довольно сложную выборку по нескольким осям одновременно
df_w.loc[i_sl[:, :, 'AC'], i_sl[[2010, 2014, 2019], :]]
```

## Пример 7.3

```python
# создаем булеву серию (по условию)
mask = df_w[(2010, 'Jan-Jun')] > 15
print(mask)
# используем сохраненную серию чтобы отфильтровать строки
df_w.loc[i_sl[mask, :, ['AC', 'TV']], :]
```
