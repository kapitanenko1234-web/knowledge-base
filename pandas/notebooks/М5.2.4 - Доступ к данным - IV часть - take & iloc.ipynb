```python
import pandas as pd

s_name = pd.Series(['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])
s_age = pd.Series([11, 23, 34, 42, 50], ['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])

d1 = {'Age':[11, 23, 34, 42, 50], 
     'Height': [140, 159, 166, 170, 182], 
     'Weight': [35, 58, 64, 68, 70]}
df1 = pd.DataFrame(d1, index=['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])
df1.index.name='Name'

d2 = {'Name':['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'],
      'Age':[11, 23, 34, 42, 50], 
      'Height': [140, 159, 166, 170, 182], 
      'Weight': [35, 58, 64, 68, 70]}
df2 = pd.DataFrame(d2)

df3 = pd.DataFrame([['Alice', 11, 140, 35],
                    ['Betty', 23, 159, 58],
                    ['Cathy', 34, 166, 64],
                    ['Dorothy',42,170, 68],
                     ['Emily', 50, 182, 70]])
d4 = {'Name':['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'],
      'Age':[11, 23, 34, 42, 50], 
      'Height': [140, 159, 166, 170, 182], 
      'Weight': [35, 58, 64, 68, 70]}
df4 = pd.DataFrame(d4, index=[1, 2, 3, 4, 5])
df4.index.name='ID'
```

# Метод .take()

```python
# достаем несколько элементов серии s_age c помощью метода .take() 
s_age.take([0,3,4])
```

```python
# достаем несколько элементов серии s_name c помощью метода .take() 
s_name.take([0,3,4])
```

```python
# запрашиваем элемент серии s_name с индексом 0 с помощью метода .take()
s_name.take([0])
```

```python
# запрашиваем элемент серии s_name с индексом 0 с помощью метода .iloc
s_name.iloc[0]
```

```python
s_name.take([-1, -2])
```

```python
# возьмем два элемента с целочисленными индексами 0 и 3 по оси строк
# параметр axis по умолчанию равен 0
# df2.take([0,3], axis=0)

df2.take([0,3])
```

```python
# возьмем два элемента с целочисленными индексами 0 и 3 по оси столбцов
# прописываем параметр axis=1

df2.take([0,3], axis=1)
```

### ПРИМЕР 3 'C'

```python
# с помощью метода iloc мы можем перезаписать строчку датафрейма
# Меняем данные Dorothy на данные Diana
df2.iloc[3] = ['Diana', 40, 172, 68]
df2
```

```python
# через метод .take() мы можем сравнить значения, но не присвоить
df2.take([3]) == ['Diana', 40, 172, 68]
df2.take([3]) == ['Dorothy', 42, 170, 68]
```
