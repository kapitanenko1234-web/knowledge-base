```python
# импортирование библиотеки pandas
import pandas as pd
```

```python
# Создаем объект DataFrame со множеством строк и столбцов в переменной df_1
age = [['George', 35, 'Male'], ['Ann', 28, 'Female'],
       ['Tom', 32, 'Male'], ['Cathy', 25, 'Female']]
df_1 = pd.DataFrame(age, columns=['Name', 'Age', 'Gender'])
df_1
```

```python
# Создаем объект DataFrame c одной строкой и тремя столбцами в переменной df_2
df_2 = pd.DataFrame([['George', 35, 'Male']], columns=['Name', 'Age', 'Gender'])
df_2
```

```python
# Создаем объект DataFrame c одной строкой и одним столбцом в переменной df_3
df_3 = pd.DataFrame(['George'], columns=['Name'])
df_3
```

```python
# Создаем объект DataFrame со множеством строк и одним столбцом в переменной df_4
df_4 = pd.DataFrame([['George'],['Ann'],['Tom'],['Cathy']], columns=['Name'])
df_4
```
