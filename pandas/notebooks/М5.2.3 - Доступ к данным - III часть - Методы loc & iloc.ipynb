### Запрашивая данные через .loc и iloc, мы можем получить:
### - Объект DataFrame 
### - Объект Series 
### - одно значение структуры, например, число

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки
# на выходе получаем серию
# Обратите внимание, что в вернувшейся серии метками-индексами являются названия столбцов датафрейма df1
df1.loc['Alice']
```

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки
# тот же результат , что и df1.loc['Alice']
df1.loc['Alice',:]
```

```python
# .loc - Обращаемся к датафрейму df1 по-метке-индексу столбца, чтобы получить данные только этого столбца
# в квадратных скобках сначала прописываем двоеточие, затем через запятую метку-индекс столбца
# на выходе получаем серию
df1.loc[:,'Age']
```

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки
# если метку-индекс поставить в еще одни квадратные скобки, то на выходе получим датафрейм с одной строкой

# Аналогичный результат дает вот такая запись, где мы дополнительно указываем полный срез по столбцам: df1.loc[['Alice'],:]

df1.loc[['Alice']]
```

```python
# .loc - Обращаемся к датафрейму df1 по-метке-индексу столбца, чтобы получить данные только этого столбца
# в квадратных скобках сначала прописываем двоеточие, затем через запятую метку-индекс столбца
# если метку-индекс столбца поставить в еще одни квадратные скобки, то на выходе получим датафрейм с одним столбцом
df1.loc[:,['Age']]
```

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки и по-метке-индексу столбца
# в квадратных скобках сначала прописываем метку-индекс строки, затем через запятую метку-индекс столбца
# на выходе получаем соответствующий значение
df1.loc['Alice','Age']
```

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки и по-метке-индексу столбца
# если мы пропишем обе метки-индексы - строки и столбца - в квадратных скобках, то на выходе получим минимальный датафрейм
df1.loc[['Alice'],['Age']]
```

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки и по срезу меток-индексов столбцов
# после запятой мы указываем выбранные нами столбцы в срезе от 'Age'до'Height' (включая 'Height'!)
# на выходе получаем серию
df1.loc['Alice','Age':'Height']
```

```python
# .loc - Обращаемся к датафрейму df1 одновременно по срезу меток-индексов строк и по срезу меток-индексов столбцов
# Последние элементы среза ВКЛЮЧАЮТСЯ
# на выходе получаем датафрейм
df1.loc['Alice':'Cathy','Age':'Height']
```

```python
# .loc - Обращаемся к датафрейму df1 по метке-индексу строки и по меткам-индексам столбцов
# после запятой мы указываем выбранные нами два столбца в списке 
# на выходе получаем серию
df1.loc['Alice',['Age','Weight']]
```

```python
# .loc - Обращаемся к датафрейму df1 по меткам-индексам строк и по меткам-индексам столбцов
# указываем элементы сначала в списке меток-строк, затем - в списке меток-столбцов
# на выходе получаем датафрейм с выбранными строками и столбцами
df1.loc[['Alice', 'Dorothy'],['Age','Weight']]
```

```python
# альтернативный вариант передачи таких же списков в метод .loc
# на выходе получаем тот же датафрейм с выбранными строками и столбцами
a = ['Alice', 'Dorothy']
b = ['Age','Weight']
df1.loc[a,b]
```

```python
# .iloc - Обращаемся к датафрейму df1 по целочисленным позициям строк и столбцов
# .iloc не используется с метками-индексами!
# Нельзя использовать синтаксис df1.iloc['Alice'], вместо этого пишем df1.iloc[0]
df1.iloc[0]
```

```python
# мы можем поместить целочисленную позицию строки '0' в еще одни квадратные скобки
# на выходе получаем датафрейм с одной строкой
df1.iloc[[0]]
```

```python
# .loc - Обращаемся к датафрейму df1 одновременно по срезу меток-индексов строк и по срезу меток-индексов столбцов
# Последние элементы среза ВКЛЮЧАЮТСЯ
df1.loc['Alice':'Cathy','Age':'Height']
```

```python
# .iloc - Обращаемся к датафрейму df1 одновременно по срезу индексов строк и по срезу индексов столбцов
# в первом срезе элементу с индексом '2' соответсвует 'Cathy', 
# во втором срезе элементу с индексом '1' соответствует 'Height'
# Последние элементы срезов НЕ ВКЛЮЧАЮТСЯ
df1.iloc[0:2,0:1]
```

```python
# .iloc - Обращаемся к датафрейму df1 одновременно по срезу индексов строк и по срезу индексов столбцов
# Если нужно, чтобы 'Cathy' и 'Height' были включены в датафрейм при использовании .iloc, мы удлиняем каждый срез на 1 элемент
df1.iloc[0:3,0:2]
```

```python
# .loc - Обращаемся к структуре df3 по целочисленному индексу, который интерпретируется как метка-индекс
df3.loc[0]
```

```python
# .iloc - Обращаемся к структуре ... по целочисленному индексу
df3.iloc[0]
```

```python
# возьмем срез с 1 по 3 строчку датафрейма df3 с помощью метода loc
# для метода .loc '1' и '3' - это проинтерпретированные метки-индексы
# Последний элемент среза ВКЛЮЧАЕТСЯ - данные Dorothy есть в выводе

df3.loc[1:3]
```

```python
# возьмем срез с 1 по 3 строчку датафрейма df3 с помощью метода iloc
# для метода .iloc '1' и '3' - это целочисленные индексы
# Последний элемент среза НЕ ВКЛЮЧАЕТСЯ - данных Dorothy нет в выводе

df3.iloc[1:3]
```

```python
# используем loc
# мы получили строчки с ID с 1 по 4, включая элемент с ID = 4
# в данном срезе указаны числовые МЕТКИ-индексы, а не индексы строк
df4.loc[1:4]
```

```python
# используем iloc
# мы получаем строчки через срез, который включил в себя значения с индексами 1,2,3 (индекс 4 не включая)
# в данном срезе указаны реальные ИНДЕКСЫ, а не ID
# В строчке с данными 'Dorothy' мы видим ID = '4', но индекс этой строки = '3'
df4.iloc[1:4]
```
