## Метод pop

Удаляет переданную колонку  из DataFrame и возвращает ее в выводе. Если метка не найдена, возвращается KeyError.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame([('group_1', 'I', 389.0),
                   ('group_1', 'II', 24.0),
                   ('group_2', 'I', 80.5),
                   ('group_2', 'II', np.nan)],
                  columns=('A', 'B', 'C'))
df
```

```python
df.pop('B')
```

```python
df
```

```python
# выдает KeyError
df.pop('D')
```
