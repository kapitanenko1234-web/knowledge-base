### Cоздаем списки фасонов и расцветок для категориальных серий

```python
# создаем полный список фасонов

type_cat = ['футляр', 'трапеция', 'годе', 'рубашка', 'маллет','бюстье', 'смок', 'с пачкой', 
            'батал', 'бейсик', 'трансформер','А-силуэт', 'миди', 'баллон', 
            'туника', 'baby-doll', 'худи', 'с запахом','лапша', 'оверсайз']

print('Всего фасонов:', len(type_cat))
```

```python
# создаем полный список расцветок

color_cat = ['красный', 'оранжевый', 'желтый', 'зеленый', 'голубой', 'синий', 'фиолетовый', 
             'сиреневый','бирюзовый','розовый','бордовый', 'коричневый', 'черный', 'белый', 
             'серый','бежевый', 'золотистый', 'серебристый', 'разноцветный']

print('Всего расцветок:', len(color_cat))
```

```python
pd.Categorical(df['Тип/Фасон'], categories=type_cat)
```

```python
pd.Categorical(df['Цвет'], categories=color_cat)
```

### Присваиваем новый объект Categorical поочередно колонке'Тип/Фасон' и колонке 'Цвет'

```python
df['Тип/Фасон'] = pd.Categorical(df['Тип/Фасон'], categories=type_cat)
```

```python
df['Цвет'] = pd.Categorical(df['Цвет'], categories=color_cat)
```

### Применяем метод value_counts к серии с типом данных 'category'

```python
df['Тип/Фасон'].value_counts()
```

```python
df['Цвет'].value_counts()
```

### Series.cat() + метод для категориального типа

    Series.cat() - Accessor object for categorical properties of the Series values.
    Объект доступа для категориальных свойств значений серии

    Series.cat.set_categories
    Series.cat.add_categories
    Series.cat.remove_categories

```python
# добавляем еще одну категорию 'малиновый' в колонку 'Цвет'
df['Цвет']=df['Цвет'].cat.add_categories(['малиновый'])
```

```python
df['Цвет'].value_counts()
```

```python
# удаляем все неиспользуемые категории в колонке 'Цвет'
df['Цвет']=df['Цвет'].cat.remove_unused_categories()
```

```python
df['Цвет'].value_counts()
```
