# Добавление столбца в DataFrame

## Использование квадратных скобок

```python
df1 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                   'B': ['B1', 'B2', 'B3'],
                   'C': ['C1', 'C2', 'C3']},
                  index=['ONE', 'TWO', 'THREE'])
display(df1)


# присваиваем существующему столбцу A значение 0
df1['A'] = 0
display(df1)
```

```python
# создаем столбец 'D'  и присваиваем ему значение 0
df1['D'] = 0
df1
```

```python
# присвоение объекта типа массива
df1['E'] = [0, 1, 2]
df1
```

```python
# ValueError: Length of values does not match length of index
df1['F'] = [0, 1, 2, 3]
```

```python
# присвоение значений серии s1 колонке 'A'

s1 = pd.Series(['A1','A2','A3'], index=['ONE', 'TWO', 'THREE'])
print(s1)
df1['A']=s1
df1
```

```python
df1['F'] = df1['B'] + df1['C']
df1['G'] = df1['B'].str.lower()
df1
```

```python
s2 = pd.Series(['X2', 'X3', 'X4'], index=['TWO', 'THREE', 'FOUR'])
print(s2)
```

```python
df1['H'] = s2
df1
```

    Атрибут values в Series возвращает массив. Элементы присваиваются по порядку, независимо от индекса. 
    Обратите внимание, что если количество элементов не совпадает с количеством строк, произойдет ошибка.

```python
s2.values
```

```python
df1['I'] = s2.values
df1
```

## Метод assign()

```python
df2 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                   'B': ['B1', 'B2', 'B3'],
                   'C': ['C1', 'C2', 'C3']},
                  index=['ONE', 'TWO', 'THREE'])
df2
```

```python
# метод assign() возвращает новый объект, при этом исходный объект остается неизменным
display(df2.assign(A=0))
display(df2.assign(D=0))
df2
```

```python
# print(s2)
df3 = df2.assign(C='XXX',
                   D=0, E=[0, 1, 2],
                   F=s2, G=s2.values,
                   H=df2['A'] + df2['B'])
df3
```

## Метод insert()

```python
df4 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                   'B': ['B1', 'B2', 'B3'],
                   'C': ['C1', 'C2', 'C3']},
                  index=['ONE', 'TWO', 'THREE'])
s3 = pd.Series(['Y2', 'Y3', 'Y4'], index=['TWO', 'THREE', 'FOUR'])

df4.insert(2, 'X', 0)
df4
```

```python
# изменения происходят в исходном объекте DataFrame
df4.insert(0, 'Y', s3)
df4
```

- Чтобы указать конец в качестве позиции для нового столбца используйте функцию len(df.columns) или метод df.shape[1] для получения количества существующих столбцов.
- Присвоение существующего имени столбца в качестве второго аргумента приведет к ошибке. Хотя можно разрешить дублирование, установив для аргумента allow_duplicates значение True

```python
df4.insert(10, 'Z', 10)
# IndexError: index 10 is out of bounds for axis 0 with size 5

# df4.insert(-1, 'Z', 10)
# ValueError: unbounded slice
```

```python
# df4.insert(0, 'Y', 10)
# ValueError: cannot insert Y, already exists
```

```python
df4.insert(0, 'Y', 10, allow_duplicates=True)
df4
```

## Функция concat() + параметр axis=1 (горизонтальное объединение)

## Пример 1 - горизонтальное объединение DataFrame и Series

```python
df5 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                   'B': ['B1', 'B2', 'B3'],
                   'C': ['C1', 'C2', 'C3']},
                  index=['ONE', 'TWO', 'THREE'])
s4 = pd.Series(['X2', 'X3', 'X4'], index=['TWO', 'THREE', 'FOUR'], name='NEW')

result_concat_1 = pd.concat([df5, s4], axis=1)
result_concat_1
```

## Пример 2 - горизонтальное объединение DataFrame и Series (только строки с общими индексами, используем параметр join='inner')

```python
# чтобы сохранить только строки с общими индексами, мы указываем параметр join='inner'
result_concat_2 = pd.concat([df5, s4], axis=1, join='inner')
result_concat_2
```

## Пример 3 - горизонтальное объединение нескольких DataFrame и Series

```python
df6 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                   'B': ['B1', 'B2', 'B3'],
                   'C': ['C1', 'C2', 'C3']},
                  index=['ONE', 'TWO', 'THREE'])
s_x = pd.Series(['X1', 'X2', 'X3'], index=df6.index, name='X')
s_y = pd.Series(['Y1', 'Y2', 'Y3'], index=df6.index, name='Y')
df7 = pd.DataFrame({'df_col1': 0, 'df_col2': range(3)}, index=df6.index)

result_concat_3=pd.concat([df6, s_x, s_y, df7], axis=1)
result_concat_3
```

```python
s5.values
```

# Объединение нескольких объектов DataFrame и Series с помощью функции concat()

```python
df9 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                    'B': ['B1', 'B2', 'B3'],
                    'C': ['C1', 'C2', 'C3']},
                   index=['ONE', 'TWO', 'THREE'])
display(df9)

df10 = pd.DataFrame({'B': ['B4', 'B5'], 'C': ['C4', 'C5'], 'D': ['D4', 'D5']},
                   index=['FOUR', 'FIVE'])
display(df10)


pd.concat([df9, df10])
```

```python
# чтобы сохранить только те столбцы, которые имеют общие имена, указываем join='inner'.

pd.concat([df9, df10], join='inner')
```

## Будьте осторожны при конкатенировании DataFrame и Series

```python
df11 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                    'B': ['B1', 'B2', 'B3'],
                    'C': ['C1', 'C2', 'C3']},
                   index=['ONE', 'TWO', 'THREE'])

s6 = pd.Series(['A4', 'B4', 'C4'], index=['A', 'B', 'C'], name='FOUR')
print(s6)

pd.concat([df11, s6])
```

## Используем метод to_frame + .T (транспонирование)

```python
# серию превращаем в датафрейм, не забыв транспонировать
display(s6.to_frame().T)

# объединяем датафреймы в один
pd.concat([df11, s6.to_frame().T])
```
