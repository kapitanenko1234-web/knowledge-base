```python
df = pd.read_csv('files/sample_1.csv', index_col=0)
df
```

# 1. Удаление строк из DataFrame

## Удаление по имени строки (по метке - label)

```python
# необходимо указать имя строки в первом аргументе labels
# аргумент axis по умолчанию стоит в позиции 0

display(df.drop('Charlie', axis=0),
        
        df.drop('Charlie'))
```

```python
# удаление строки с использованием аргумента index

df.drop(index='Charlie')
```

```python
# используем список, в который помещаем все метки строк под удаление

display(df.drop(['Bob', 'Dave', 'Frank']),
        
        df.drop(index=['Bob', 'Dave', 'Frank']))
```

```python
df_1 = df.copy()
# если аргумент inplace=True, изменяется исходная структура DataFrame и ничего не возвращается (None):

print(df_1.drop(index=['Bob', 'Dave', 'Frank'], inplace=True))
```

```python
# строк с метками 'Bob', 'Dave', 'Frank' блольше нет в структуре df_1
df_1
```

```python
df_2 = pd.read_csv('files/sample_1.csv', index_col=0)
df_2
```

```python
df_add = pd.DataFrame({'age': [20, 18, 25], 'state': ['NY','TX','CA'], 'point':[80, 76, 89]}, 
                      index=['Ellen', 'George', 'Dave'])
df_add                                        
```

```python
df_new=pd.concat([df_2, df_add])
df_new
```

```python
# удаляются все строчки с метками 'Dave' и 'Ellen'

df_new.drop(['Dave', 'Ellen'])
```

```python
# Обнаружить дубликаты помогает метод duplicated()
# метод duplicated() возвращает булев массив
# с помощью метода sum() можно подсчитать количество дубликатов

print(df_new.index.duplicated(),
      'Количество дубликатов:', df_new.index.duplicated().sum(), sep='\n\n')
```

## Удаление по номеру строки

```python
df
```

```python
# мы можем использовать атрибут index в качестве первого аргумента labels или в качестве аргумента index метода drop()

display(df.drop(df.index[[1, 3, 5]]),
        df.drop(index=df.index[[1, 3, 5]]))
```

## Если индекс не задан

```python
df_no_index = pd.read_csv('files/sample_1.csv')

display(df_no_index)

print('Индексы структуры df_no_index:', df_no_index.index)
```

```python
# В данном примере прямое указание номеров строк при удалении дает тот же результат, что и использование атрибута index.

display(df_no_index.drop([1, 3, 5]))

display(df_no_index.drop(df_no_index.index[[1, 3, 5]]))
```

```python
# Результат будет отличаться, если последовательность нарушена такими действиями, как сортировка (sort_values('state'))

df_no_index_sort = df_no_index.sort_values('state')
display(df_no_index_sort)

display(df_no_index_sort.index)
```

```python
# При прямом указании числа удаляется строка с этим числом в качестве метки

display(df_no_index_sort.drop([1, 3, 5]))

display(df_no_index_sort.drop(df_no_index_sort.index[[1, 3, 5]]))
```

# 2. Удаление столбцов из DataFrame

## Удаление по имени столбца (по метке)

```python
df.drop('state', axis=1)
```

```python
df.drop(columns='state')
```

```python
display(df.drop(['state', 'point'], axis=1),
        df.drop(columns=['state', 'point']))
```

## Удаление по номеру столбца

```python
display(df.columns[[1, 2]])

# используем атрибут columns структуры DataFrame

display(df.drop(df.columns[[1, 2]], axis=1),
        
        df.drop(columns=df.columns[[1, 2]]))
```

## Одновременное удаление нескольких строк и столбцов

```python
# используем одновременно аргументы index и columns

df.drop(index=['Bob', 'Dave', 'Frank'], columns=['state', 'point'])
```

```python
# используем одновременно аргументы index и columns + атрибуты index и columns

df.drop(index=df.index[[1, 3, 5]], columns=df.columns[[1, 2]])
```

```python
df_copy = df.copy()

df_copy.drop(index=['Bob', 'Dave', 'Frank'], columns=['state', 'point'],
             inplace=True)
df_copy
```
