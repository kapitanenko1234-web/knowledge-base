```python
# импортирование библиотеки pandas
import pandas as pd

# Создаем объект Series в переменной s_1
# индексы - по умолчанию представлены объектом RangeIndex (монотонный диапазон целых чисел)
# имя объекта Series не задано
# тип данных Series определен автоматически (int64) в момент создания объекта
s_1 = pd.Series(data=['a', 'b', 'c', 'd', 'e'])
s_1
```

```python
# Создаем объект Series в переменной s_2
# задаем index (в результате получаем метки-индексы (labels))
# задаем имя объекта Series
# тип данных Series определен автоматически (int64) в момент создания объекта
s_2 = pd.Series(data=[1,3,5,7,9], index=['one','three','five','seven','nine'], name='odd_numbers')
s_2
```

```python
# Создаем объект DataFrame в переменной df
age = [['George', 35, 'Male'], ['Ann', 28, 'Female'],
       ['Tom', 32, 'Male'], ['Cathy', 25, 'Female']]
df = pd.DataFrame(age, columns=['Name', 'Age', 'Gender'])
```

```python
df
```

```python
# здесь объект Series - это колонка структуры DataFrame
# данные структуры Series однотипные (int64) - весь ряд представлен в виде целых чисел (int64)
# имя структуры Series - 'Age' - это имя запрошенной колонки структуры DataFrame в переменной df
df['Age']
```

```python
df
```

```python
# здесь объект Series - это строка структуры DataFrame
# данные структуры Series разнотипные (object):
# метка 'Name' - строковое значение
# метка 'Age' - целое число
# метка 'Gender' строковое значение
# имя серии - 1 (целочисленная позиция строки 1 структуры DataFrame в переменной df)
df.iloc[1]
```

```python
# доступ к элементу структуры Series по целочисленной позиции (1)
s_2.iloc[1]
```

```python
# доступ доступ к элементу структуры Series по метке ('three')
s_2.loc['three']
```
