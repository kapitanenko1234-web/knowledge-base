## columns.str.replace

## Изменение имен столбцов с помощью skimpy

### https://aeturrell.github.io/skimpy/

### pip install skimpy

```python
from skimpy import clean_columns
```

```python
columns = ['_Customer__name',
           'customer_génder',
           'customer age',
           'billingAddress',
           'delivery** address  ',
           'item(Description)',
           'quantity/Volume',
           'total.price']
messy_df = pd.DataFrame(data = [], columns=columns, index=[0])
messy_df
```

### 1. Snake Case

### Стиль Snake Case заменяет пробелы и разделители символов подчеркиванием и преобразует все символы в нижний регистр.

```python
clean_df1 = clean_columns(messy_df)
clean_df1
```

### 2. Kebab Case

### Стиль kebab case аналогичен snake, за исключением того, что вместо подчеркивания используется разделитель тире.

```python
clean_df2 = clean_columns(messy_df, case = 'kebab')
clean_df2
```

### 3. Camel Case

### Cтиль Camel case использует символы верхнего регистра в качестве разделителя, а первый символ каждой строки находится в нижнем регистре.

```python
clean_df3 = clean_columns(messy_df, case = 'camel')
clean_df3
```

### 4. Pascal Case

### Стиль pascal подобен  Camel Case, за исключением того, что первый символ каждой строки находится в верхнем регистре.

```python
clean_df4 = clean_columns(messy_df, case = 'pascal')
clean_df4
```

### 5. Constant Case

### Стиль constant case использует разделитель подчеркивания и изменяет все символы на верхний регистр.

```python
clean_df5 = clean_columns(messy_df, case = 'const')
clean_df5
```

## Замена названия с помощью clean_columns()

```python
# сокращаем Address до Addr и Description до Desc

clean_df6 = clean_columns(messy_df, case = 'pascal', replace = {'Address':'Addr', 'Description':'Desc'})
clean_df6
```
