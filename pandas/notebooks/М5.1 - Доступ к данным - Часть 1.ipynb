```python
import pandas as pd
```

```python
# Создаем серию даных (метки-индексы - женские имена, значения - возраст) 
s_age = pd.Series([11, 23, 34, 42, 50], ['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])
s_age
```

```python
# Создаем DataFrame с параметрами женщин - Age, Height, Weight - Возраст, Рост, Вес
# Метки-индексы строк - те же женские имена (по первым пяти буквам английского алфавита)
# Даем колонке с индексами название 'Name'
d = {'Age':[11, 23, 34, 42, 50], 'Height': [140, 159, 166, 170, 182], 'Weight': [35, 58, 64, 68, 70]}
df = pd.DataFrame(d, index=['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])
df.index.name='Name'
df
```

# Атрибуты .INDEX, .VALUES, .COLUMNS, .AXES

```python
# достаем все индексы серии
s_age.index
```

```python
# достаем все значения серии
s_age.values
```

```python
# Еще раз взгляем на датафтрейм
df
```

```python
# Получаем доступ ко всем индексам СТРОК датафрейма
df.index
```

```python
# Получаем доступ ко всем индексам СТОЛБЦОВ датафрейма
df.columns
```

```python
# Получаем доступ ко всем ЗНАЧЕНИЯМ датафрейма
df.values
```

```python
# если запросить колонку датафрейма вернется серия со значениями колонки, именем серии будет название колонки
df['Age']
```

```python
# с помощью атрибута .values мы можем запросить значения колонки списком (как значения серии)
df['Age'].values
```

```python
# с помощью атрибута .axes запрашиваем массив с двумя списками - индексы СТРОК и СТОЛБЦОв датафрейма
df.axes
```

```python
# .axes можно запросить и у серии, но в массиве будет только список с идексами СТРОК
s_age.axes
```

```python
# DataFrame c индексами, проставленными по умолчанию (RangeIndex)
df2=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]])
df2
```

```python
# RangeIndex - Неизменяемый индекс,реализующий монотонный целочисленный диапазон.
df2.index
```

```python
df2.columns
```

```python
df2.axes
```
