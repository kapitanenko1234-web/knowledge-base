```python
from glob import glob
```

#### Вы можете передать функции glob() шаблон, включая символы подстановки, и он вернет список всех файлов, соответствующих этому шаблону.

#### В данном случае glob ищет в подкаталоге "files" все CSV-файлы, начинающиеся со слова "stocks"

```python
stock_files = sorted(glob('files/stocks*.csv'))
stock_files
```

glob возвращает имена файлов в произвольном порядке, поэтому мы отсортировали список с помощью встроенной в Python функции sorted().

Затем мы можем использовать выражение-генератор для чтения каждого из файлов с помощью read_csv() и передать результаты в функцию concat(), которая объединит строки в один DataFrame:

```python
pd.concat((pd.read_csv(file) for file in stock_files))
```

## Построение DataFrame из нескольких файлов (объединение по столбцам)

```python
pd.concat((pd.read_csv(file) for file in stock_files), ignore_index=True)
```
