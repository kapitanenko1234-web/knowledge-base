```python
import pandas as pd
import numpy as np
```

```python
# cоздадим серию s3 с оценками студентов
grades = [[5, 4, 3], 
      [4, 3, 2], 
      [3, 5, 5],
      [3, 3, 3],
      [5, 5, 4]]
students = ['student1@mail.ru',
          'student2@mail.ru',
          'student3@mail.ru',
          'student4@mail.ru',
          'student5@mail.ru']
s3 = pd.Series(grades, students)
s3
```

```python
# передаем s3 в коструктор DataFrame
df_students = pd.DataFrame(s3)
print(type(df_students))
df_students
```

```python
# используем метод to_frame()
df_students_ = s3.to_frame()
df_students_
```

# 1 способ

### Пример № 1

```python
# посмотрим на списки индексов и значений структуры s3
print(s3.index)
print(s3.values)
```

```python
# используем атрибуты index и values как значения словаря, ключи словаря - названия колонок 
# словарь передаем в конструктор DataFrame
# используем атрибуты index и name, чтобы дать имя индексам
df_students1 = pd.DataFrame({'e-mail':s3.index, 'grades':s3.values})
df_students1.index.name = 'index'
df_students1
```

# 2 способ 

### Пример № 2.1

```python
# используем:
# to_frame() + reset_index() 
# атрибут columns, чтобы дать имена колонкам 
# атрибуты index и name, чтобы дать имя индексам
df_students2_1 = s3.to_frame().reset_index()
df_students2_1.columns =['e-mail', 'grades']
df_students2_1.index.name = 'index'
df_students2_1
```

### Пример № 2.2

```python
# используем:
# to_frame() + reset_index() 
# метод .rename чтобы переименовать колонки 
# атрибуты index и name, чтобы дать имя индексам
df_students2_2 = s3.to_frame().reset_index()
df_students2_2 = df_students2_2.rename(columns={"index":"почта студента", 0:"оценки"})
df_students2_2.index.name = 'индекс'
df_students2_2
```
