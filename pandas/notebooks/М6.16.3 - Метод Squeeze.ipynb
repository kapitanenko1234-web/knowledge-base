## Метод Squeeze

```python
import pandas as pd
```

```python
numbers = pd.Series([2, 3, 5, 7])
numbers
```

```python
even_numbers = numbers[numbers % 2 == 0]
even_numbers
```

```python
even_numbers.squeeze()
```

```python
# Сжатие объектов с более чем одним значением по каждой оси ничего не дает:
odd_numbers = numbers[numbers % 2 == 1]
odd_numbers
```

```python
odd_numbers.squeeze()
```

```python
# Сжатие становится еще более эффективным при использовании со структурой DataFrame
df = pd.DataFrame([[1, 2], [3, 4]], columns=['a', 'b'])
df
```

```python
# При доступе к одному столбцу получается DataFrame, в котором строки имеют только одно значение:
df_a = df[['a']]
df_a
```

```python
# Поэтому колонку можно сжать, в результате чего получится серия:
df_a.squeeze('columns')
```

```python
# При нарезке одной строки из одного столбца будет получен один скалярный DataFrame:
df_0a = df.loc[df.index < 1, ['a']]
df_0a
```

```python
# При сжатии рядов получается одна скалярная серия:
df_0a.squeeze('index')
```

```python
# При сжатии столбцов получается также одна скалярная серия:
df_0a.squeeze('columns')
```

```python
df_0a.squeeze()
```
