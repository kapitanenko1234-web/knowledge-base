# Timestamp & Timedelta

```python
pd.Timestamp('25/12/2019')
```

```python
# различные форматы входных данных для создания объекта Timestamp
display(pd.Timestamp('25 December 2019'),
      pd.Timestamp('December 25 2019'),
      pd.Timestamp('12/25/2019'),
      pd.Timestamp('25-12-2019'),
      pd.Timestamp(year=2019,month=12,day=25),
      pd.Timestamp('25-12-2019 0 AM'),
      pd.Timestamp('25-12-2019 0:0.0'))
```

#### отдельная функция, pd.Timedelta, помогает нам создавать объекты, хранящие длительность времени. 

```python
pd.Timedelta('45 days 9 minutes')
```

```python
pd.Timedelta('45 days 00:09:00')
```

```python
pd.Timedelta(days=45,minutes=9)
```

```python
# мы также можем добавить параметр unit, чтобы создать объект Timedelta.
pd.Timedelta(500,unit='m')
```

## Преобразование строк в объекты Pandas Timestamp

```python
a = pd.to_datetime('11:20,02/04/1990', format='%H:%M,%d/%m/%Y')
a
```

```python
# к объекту Timestamp можно добавить объект Timedelta. Давайте добавим к этой дате четыре дня:
a + pd.Timedelta(4,unit='d')
```
