```python
import pandas as pd
import numpy as np


d2 = {'Name':['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'],
      'Age':[11, 23, 34, 42, 50], 
      'Height': [140, 159, 166, 170, 182], 
      'Weight': [35, 58, 64, 68, 70]}
df2 = pd.DataFrame(d2)
```

# Дополнительные возможности loc и iloc

## Доступ к данным (loc и iloc + булев массив)

```python
df7=df2.copy()
df7
```

```python
# В переменной 'a' содержится массив из пяти элементов (в соответствии с количеством строк датафрейма df7)
a=[True, False, True, False, True]
# В переменной 'b' содержится массив из четырех элементов (в соответствии с количеством столбцов датафрейма df7)
b=[True, False, False, True]
```

```python
# используем метод loc c булевыми массивами
df7.loc[a, b]
```

```python
# используем метод iloc c булевыми массивами
df7.iloc[a, b]
```

```python
# фильтруем только столбцы с помощью loc и булева массива 
# соответсвует записи df7.loc[:,b]

df7.loc[:,[True, False, False, True]]
```

```python
# фильтруем только строки с помощью iloc и булева массива 
# соответсвует записи df7.iloc[a]

df7.iloc[[True, False, True, False, True]]
```

# Возможность изменения структуры

## Пример 1.1

```python
df8=df2.copy()
df8
```

```python
df8.loc[2,'Weight']
```

```python
df8.loc[2,'Weight']=66
df8
```

```python
# мы также можем использовать iloc для изменения структуры, используя для доступа к данным целочисленные позиции элементов
df8_2=df2.copy()
df8_2.iloc[2, 3]=66
df8_2
```

## Пример 1.2

```python
df8.loc[5]=['Felicity', 56, 183, 75]
df8
```

```python
df8.loc[6,'Name']='Grace'
df8
```

## Пример 2

```python
df9=df2.copy()
df9
```

```python
df9.loc[:,'Weight']-1
```

```python
# если не присвоить результат вычисления соответствующей колонке, то датафрейм df9 останется прежним
df9
```

```python
# присваиваем результат вычисления соотвествующей колонке датафрейма df9
df9.loc[:,'Weight']=df9['Weight']-1
df9
```

## Пример 3.1

#### Индекс массы тела (ИМТ) - body mass index (BMI)
#### Для вычислений необходимо взять вес в килограммах и разделить на рост в метрах, возведенный в квадрат.

```python
# мы одновременно создаем колонку и производим вычисления, чтобы заполнить эту колонку значениями
df10=df9.copy()
df10.loc[:, 'BMI'] = (df10['Weight']/((df10['Height']/100)**2)).round(1)
df10
```

## Пример 3.2 

```python
df10_2=df9.copy()
df10_2['BMI'] = (df10_2['Weight']/((df10_2['Height']/100)**2)).round(1)
df10_2
```

# Возможность одновременного поиска и создания колонки по условию

```python
# Пример поиска по условию и одновременного создания колонки со значениями через loc
# Сначала создаем копию датафрейма df2 - df11
df11=df2.copy()
df11
```

```python
# Ищем все строки, где возраст менее 18 лет, добавляем новую колонку 'Child' и сразу присваиваем значение 1:
df11.loc[df11['Age']<18, 'Child']=1
df11
```

```python
df11['Child'].dtype
```

```python
# incompatible dtype
df12=df2.copy()
df12.loc[df12['Age']<18, 'Child']='yes'
df12
```

```python
df13=df2.copy()
df13.loc[:, 'Child']=None
df13
```

```python
df13['Child']
```

```python
df13['Child'].dtype
```

```python
df13.loc[df13['Age']<18, 'Child']='yes'
df13
```

```python
# для красоты заменим все отсутсвующие значения в колонке 'Child' на 'no':
df13['Child'].fillna('no', inplace=True)
df13
```

# Другие методы pandas 

```python
df14=df2.copy()
df14['Child'] = np.where(df14['Age'] < 18, 'yes', 'no')
df14
```

```python
df15=df2.copy()
df15['Age_category'] = pd.cut(df15['Age'], [0,18,120], labels=['child', 'adult'], right=False)
df15
```
