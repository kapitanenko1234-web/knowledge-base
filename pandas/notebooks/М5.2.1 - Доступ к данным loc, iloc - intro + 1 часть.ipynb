# I часть - Подготавливаем серии и датафреймы

## Серия s_name

```python
# Создаем первую серию данных (значения - женские имена, целочисленные индексы появляются автоматически по умолчанию)
s_name = pd.Series(['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])
s_name
```

## Серия s_age

```python
# Создаем вторую серию даных (метки-индексы - женские имена, значения - возраст) 
s_age = pd.Series([11, 23, 34, 42, 50], ['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])
s_age
```

_Создадим четыре структры DataFrame, в которых будут фигурировать данные о 5 девушках - Age, Height, Weight - Возраст, Рост, Вес_

## DataFrame df1

```python
# Создаем DataFrame df1 

# индексы строк- метки-индексы (labels - 'Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily')
# индексы столбцов - метки-индексы (labels - 'Age', 'Height', 'Weight')

d1 = {'Age':[11, 23, 34, 42, 50], 
     'Height': [140, 159, 166, 170, 182], 
     'Weight': [35, 58, 64, 68, 70]}
df1 = pd.DataFrame(d1, index=['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'])

# Даем колонке с индексами название 'Name'
df1.index.name='Name'

df1
```

## DataFrame df2

```python
# Создаем DataFrame df2

# индексы строк- целочисленные значения RangeIndex
# индексы столбцов - метки-индексы (labels - 'Name', 'Age', 'Height', 'Weight')

d2 = {'Name':['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'],
      'Age':[11, 23, 34, 42, 50], 
      'Height': [140, 159, 166, 170, 182], 
      'Weight': [35, 58, 64, 68, 70]}
df2 = pd.DataFrame(d2)

df2
```

_Еще один способ создать датафрейм df2 это использовать датафрейм df1 вместе с reset_index_

## DataFrame df3

```python
# создаем еще один DataFrame df3 с такими же данными, только без меток-индексов
# обратите внимание, что мы передаем в конструктор класса не словарь, а двумерный массив

# индексы строк - целочисленные значения RangeIndex
# индексы столбцов - целочисленные значения RangeIndex

df3 = pd.DataFrame([['Alice', 11, 140, 35],
                    ['Betty', 23, 159, 58],
                    ['Cathy', 34, 166, 64],
                    ['Dorothy',42,170, 68],
                     ['Emily', 50, 182, 70]])
df3
```

## DataFrame df4

```python
# Создаем DataFrame df4

# индексы строк- метки-индексы, в качестве которых выступают целые цисла (labels - 1, 2, 3, 4, 5)
# индексы столбцов - метки-индексы (labels - 'Name', 'Age', 'Height', 'Weight')

d4 = {'Name':['Alice', 'Betty', 'Cathy', 'Dorothy', 'Emily'],
      'Age':[11, 23, 34, 42, 50], 
      'Height': [140, 159, 166, 170, 182], 
      'Weight': [35, 58, 64, 68, 70]}
df4 = pd.DataFrame(d4, index=[1, 2, 3, 4, 5])

# Даем колонке с индексами название 'ID'
df4.index.name='ID'

df4
```
