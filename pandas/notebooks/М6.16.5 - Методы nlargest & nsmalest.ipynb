```python
import pandas as pd
```

```python
df_movies=pd.read_csv('files/movies_1000', usecols=['star_rating','title','genre'])
df_movies.head(10)
```

Нам нужно отфильтровать DataFrame по жанру, но включить в него только 3 самых крупных жанра.

Начнем с того, что получим значения value_counts() для genre и сохраним его в виде серии под названием vc:

```python
vc = df_movies['genre'].value_counts()
vc
```

Метод Series.nlargest() позволяет легко выбрать 3 самых больших значения в этой серии:

И все, что нам нужно от этой серии, - это индекс:

```python
vc.nlargest(3)
```

```python
vc.nlargest(3).index
```

Наконец, мы можем передать этот объект index в метод isin() и по полученному булеву массиву произвести фильтрацию структуры:

```python
df_movies[df_movies['genre'].isin(vc.nlargest(3).index)].head()
```

#### находим 3 самых малочисленных жанра с методом nsmallest

```python
df_movies[df_movies['genre'].isin(vc.nsmallest(3).index)].head()
```

```python
vc.nsmallest(3).index
```
