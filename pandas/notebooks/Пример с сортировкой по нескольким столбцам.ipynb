```python
import numpy as np
import pandas as pd
```

```python
df_new = pd.read_csv('files/sample_1.csv')
df_new
```

# сразу два шага сортировки в одной строчке кода выглядят так
#### сохраняем результат в переменную sorted_df

```python
sorted_df=df_new.sort_values(['age', 'state'], ascending=[True, False])
sorted_df
```

# но мы можем сделать то же самое в 2 этапа для наглядности
#### используем inplace=True, чтобы сохранить в переменной df_new результат 1-й сортировки

### 1 этап сортировки

```python
df_new.sort_values(['state'], ascending=[False], inplace=True)
df_new
```

### 2 этап сортировки

```python
result = df_new.sort_values(['age'], ascending=[True])
result
```

#### Конечный результат второй сортировки сохраняем в переменную result и сравниваем с датафреймом в sorted_df

```python
sorted_df == result
```
