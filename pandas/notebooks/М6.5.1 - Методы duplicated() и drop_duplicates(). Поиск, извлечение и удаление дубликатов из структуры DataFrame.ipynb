    План урока
    
    1. Поиск, извлечение и подсчет дублирующихся строк: duplicated()
    - Базовое использование метода duplicated()

    - Указание столбцов для обнаружения дубликатов: аргумент subset
    - Подсчет дублирующихся и недублирующихся строк
    
    2. Удаление дубликатов: drop_duplicates()

    - Выбор дубликатов для сохранения: аргумент keep
    - Указание столбцов для обнаружения дубликатов: аргумент subset
    - Сброс индекса: аргумент ignore_index
    - Изменение исходного объекта: аргумент inplace

```python
df = pd.read_csv('files/sample_1.csv')
df.loc[6] = ['Dave', 68, 'TX', 70]
df
```

## Базовое использование метода duplicated()

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

## Выбор дубликатов для сохранения: аргумент keep


```python
# По умолчанию (keep='first') отмечает все дубликаты строк как True, за исключением первого, который отмечается как False

display(df.duplicated(),
        # то же самое:
        df.duplicated(keep='first'))
```

```python
# при установке keep='last' все дубликаты, кроме последнего, помечаются как True

df.duplicated(keep='last')
```

```python
# установка keep=False отмечает все дубликаты как True

df.duplicated(keep=False)
```

## Указание столбцов для обнаружения дубликатов: аргумент subset

```python
df
```

```python
df.duplicated(subset='state')
```

```python
df.duplicated(subset=['state', 'point'])
```

## Подсчет дублирующихся и недублирующихся строк

```python
# вычисляем количество дубликатов

df.duplicated().sum()
```

```python
# находим недублирующиеся строки

~df.duplicated()
```

```python
# вычисляем количество недублирующихся строк

(~df.duplicated()).sum()
```

```python
# вычисляем количество дублирующихся и недублирующихся строк одновременно

df.duplicated().value_counts()
```

```python
df.duplicated(keep=False).value_counts()
```

## Удаление дубликатов строк: drop_duplicates()

## Базовое использование метода drop_duplicates()

```python
df.drop_duplicates()
```

```python
df
```

## Выбор дубликатов для сохранения: аргумент keep

- keep='first' (по умолчанию) сохраняет ПЕРВОЕ вхождение каждой дублирующейся строки
- keep='last' сохраняет ПОСЛЕДНЕЕ вхождение каждой дублирующейся строки
- Установка keep=False удаляет ВСЕ ВХОЖДЕНИЯ ДУБЛИКАТОВ

```python
df.drop_duplicates(keep='last')
```

```python
df.drop_duplicates(keep=False)
```

## Указание столбцов для обнаружения дубликатов: аргумент subset

```python
df.drop_duplicates(subset='state')
```

```python
df.drop_duplicates(subset=['state', 'point'])
```

## Сброс индекса: аргумент ignore_index

```python
df.drop_duplicates(subset='state', keep='last')
```

```python
df.drop_duplicates(subset='state', keep='last', ignore_index=True)
```

## Изменение исходного объекта: аргумент inplace

```python
df.drop_duplicates(subset='state', keep='last', inplace=True)
df
```
